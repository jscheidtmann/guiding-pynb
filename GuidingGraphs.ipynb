```python
import matplotlib.pyplot as plt
import numpy as np
import math

worm_period = 480.
periods = 3
sampling = 1.

frequencies = [480., 120.]
amplitudes = [13, 1.6]
phases = np.random.uniform(size=len(frequencies))

x = np.arange(0., periods*worm_period, sampling)
y = np.zeros(len(x))
for i in range(len(frequencies)): 
    y = y + amplitudes[i]*np.sin(math.pi * 2. / frequencies[i] * x + math.pi * 2. * phases[i])
    plt.plot(x, amplitudes[i]*np.sin(math.pi * 2. / frequencies[i] * x + math.pi * 2. * phases[i]))

plt.savefig("frequenzen.png", transparent=True)

```

```python
plt.plot(x,y, "r-")

plt.savefig("frequenzen_combined.png", transparent=True)
```

```python
from scipy.fftpack import fft, ifft

X = fft(y)

plt.figure(figsize = (12, 6))
plt.subplot(121)

plt.plot(np.abs(X))
# plt.stem(freq, np.abs(X), 'b', \
#         markerfmt=" ", basefmt="-b")
plt.xlabel('Freq (Hz)')
plt.ylabel('FFT Amplitude |X(freq)|')
plt.xlim(0, 200)

plt.subplot(122)
plt.plot(y, ifft(X), 'r')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.tight_layout()
plt.show()
```
